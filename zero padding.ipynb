# zero padding

```python
N = 3
arr = [ [0] + list(map(int,input().split())) + [0] for _ in range(N) ]
print(arr)
```

---

```python
N = 3
arr = []
    
for _ in range(N):
    arr.append(list(map(int,input().split())))
        

# zero padding 
for lst in arr:
    lst.insert(0,0)
    lst.append(0)

print(arr)
```
