# List 입력

```python
lst = [[0 for j in range(2)] for i in range(3)]
lst
```

```python
lst = [[0] * 2 for i in range(3)]
lst
```

```python
lst = []    # 빈 리스트 생성
 
for i in range(3):
    lst_in = []              # 안쪽 리스트로 사용할 빈 리스트 생성
    
    for j in range(2):
        lst_in.append(0)     # 안쪽 리스트에 0 추가
    
    lst.append(lst_in)       # 전체 리스트에 안쪽 리스트를 추가
    
print(lst)
```

```python
pprint(lst, indent=4, width=20)
```

## sorted로 2차원 리스트 정렬하기

- sorted(반복가능한객체, key=정렬함수, reverse=True 또는 False)

- 다음은 학생 정보가 저장된 2차원 리스트를 정렬합니다.

```python
students = [
    ['john', 'C', 19],
    ['maria', 'A', 25],
    ['andrew', 'B', 7]
]
 
print(sorted(students, key=lambda student: student[1]))  # 안쪽 리스트의 인덱스 1을 기준으로 정렬
print(sorted(students, key=lambda student: student[2]))  # 안쪽 리스트의 인덱스 2를 기준으로 정렬
```

# 2D List 출력

## `for x,y in lst`

```python
lst = [[10, 20], [30, 40], [50, 60]]
```

```python
for x, y in lst:    # 리스트의 가로 한 줄(안쪽 리스트)에서 요소 두 개를 꺼냄
    print(x,y)
```

```python
for i in lst:
    print(i[0],i[1])
```

## Nested for loop

```python
lst = [[10, 20], [30, 40], [50, 60]]
 
for i in lst:        # 전체 리스트에서 가로 한 줄씩 꺼내 옵니다(안쪽 리스트를 통째로 꺼내 옴). 
    for j in i:      # 다시 한 줄(안쪽 리스트) i에서 요소 j를 하나씩 꺼내면 됩니다.
        print(j, end=' ')
    print()
```

## for range len

```python
lst = [[10, 20], [30, 40], [50, 60]]
lst
```

```python
from pprint import pprint

# indent는 들여쓰기 칸 수, width는 가로 폭입니다. 

pprint(lst, indent=4, width=20)
```

```python
for i in range(len(lst)):            # 전체 리스트 크기 : 세로 크기
    for j in range(len(lst[i])):     # 내부 리스트 : 가로 크기
        print(lst[i][j], end=' ')
    print()
```

## while

```python
lst = [[10, 20], [30, 40], [50, 60]]
```

```python
i = 0
while i < len(lst):    # 반복할 때 리스트의 크기 활용(세로 크기)
    x, y = lst[i]      # 요소 두 개를 한꺼번에 가져오기
    print(x, y)
    i += 1           # 인덱스를 1 증가시킴
    
```

```python
lst = [[10, 20], [30, 40], [50, 60]]
 
i = 0
while i < len(lst):           # 세로 크기
    
    j = 0
    
    while j < len(lst[i]):    # 가로 크기
        print(lst[i][j], end=' ')
        j += 1              # 가로 인덱스를 1 증가시킴
    
    print()
    i += 1                  # 세로 인덱스를 1 증가시킴
```
